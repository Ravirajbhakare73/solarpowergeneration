```python
#Bivariate analysis 
plt.figure(figsize=(15,30))
for i, column in enumerate(dataframe.columns):
    plt.subplot(7,3,i+1)
    plt.scatter(dataframe[column], dataframe['generated_power_kw'])

    plt.title(f'{column} vs Generated Power in kw')
    plt.xlabel(column)
    plt.ylabel("Generated Power (kw)")

plt.tight_layout()
plt.show()
    
```
